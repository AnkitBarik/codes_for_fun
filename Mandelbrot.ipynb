```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
from pylab import *
```

```python
N=1000
x = linspace(-2.,2.,N)
y = linspace(-2.,2.,N)

X,Y = meshgrid(x,y)

c = X + 1j*Y
cM = zeros_like(c)
col = zeros(shape(c))
```

```python
def inMand(c):
    z = 0
    count = 0
    
    for i in range(15):
        z = z**2 + c
        if (abs(z) > 5):
            return False
    return True

def inMand_col(c):
    z = 0
    count = 0
    
    for i in range(15):
        z = z**2 + c
        count +=1
        if (abs(z) > 5):
            return count
    return count
```

```python
for i in range(N):
    for j in range(N):
        if(inMand(c[i,j])):
            cM[i,j] = c[i,j]

cM[abs(cM) > 0] = 1.
figure(figsize=(6,6))
contourf(x,y,abs(cM),cmap='Greys')
show()
```

```python
for i in range(N):
    for j in range(N):
        col[i,j] = inMand_col(c[i,j])

figure(figsize=(6,6))
contourf(x,y,col,cmap='plasma')
show()
```
